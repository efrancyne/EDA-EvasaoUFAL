#### Organizando e limpando os dados

```python
df.isna().sum()
```

```python
df.head()
```

```python
df = df.dropna(subset=['idade'])
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df = df.drop(['unica'], axis=1)
```

```python
df.columns
```

```python
cursos = df['curso'].str.split(' - ', expand=True)
```

```python
df['nome_curso'] = cursos[0]
df['campus'] = cursos [1]
df['turno'] = cursos [2]
df['modalidade'] = cursos [3]
df['grau'] = cursos [4]
df['regime'] = cursos [5]
df['codigo_curso'] = cursos [6]
```

```python
df.head(3)
```

```python
df.columns
```

```python
df_organizado = df[
    [

        # Informações do curso
        'nome_curso',
        'grau',
        'turno',
        'modalidade',
        'regime',
        
        # Perfil do aluno
        'genero',
        'idade',
        'etnia',

        # Localização
        'campus',
        'unidade',
        # Tempo
        'periodo_letivo',

    ]
]
```

```python
dados_ingressantes = df_organizado
```

```python
dados_ingressantes.head()
```
