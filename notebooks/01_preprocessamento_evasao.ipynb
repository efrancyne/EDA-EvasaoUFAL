```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

### Carregando os dados

```python
df_evasao = pd.read_csv("../dados/dados_brutos/dados_evasao.csv")
```

```python
df_evasao.head()
```

### Limpeza e organização dos dados

```python
df_evasao.isna().sum()
```

```python
df_evasao['idade'].describe
```

```python
df_evasao = df_evasao.dropna(subset=['idade'])
```

```python
df_evasao['necessidade_especial'].value_counts(dropna=False)
```

```python
df_evasao['necessidade_especial'] = (
    df_evasao['necessidade_especial']
    .fillna('Nenhuma')
)
```

```python
div_var_cursos = df_evasao['curso'].str.split(' - ', expand=True)
```

```python
df_evasao['nome_curso'] = div_var_cursos[0]
df_evasao['campus'] = div_var_cursos [1]
df_evasao['turno'] = div_var_cursos [2]
df_evasao['modalidade'] = div_var_cursos [3]
df_evasao['grau'] = div_var_cursos [4]
df_evasao['regime'] = div_var_cursos [5]
```

```python
df_evasao_processado = df_evasao[
    [

        # Informações do curso
        'nome_curso',
        'grau',
        'turno',
        'modalidade',
        'regime',
        'nivel',
        
        # Perfil do aluno
        'genero',
        'idade',
        'etnia',
        'necessidade_especial',

        # Localização
        'campus',
        'unidade',
        'sigla_unidade',

        # Situação acadêmica
        'status',
        'forma_evasao',
        'tipo_evasao',

        # Tempo
        'periodo_letivo',

    ]
]
```

```python
df_evasao_processado.to_csv("../dados/dados_preprocessados/evasao_ufal.csv")
```
